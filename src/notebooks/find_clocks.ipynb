## rosetta?

```python
clock_ids = set()
with gz.open('../data/rosetta/cells.json.gz', 'rt') as f:
    data = json.load(f)

for cell in data:
    if 'clk' in cell['name'].lower():
        print(cell)
        clock_ids.add(cell['id'])
```

```python
clk = []
for design in range(1,20):
    print(design)
    if design in [8,13,17]:
        continue
    with gz.open(f'../data/rosetta/superblue{design}/superblue{design}.json.gz', 'rt') as f:
        data = json.load(f)
    
    for instance in data['instances']:
        if instance['cell'] in clock_ids:
            print(instance)
            clk.append(instance)
        if 'clk' in instance['name'].lower():
            print(instance)
```
